```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_gen=ImageDataGenerator(zoom_range=0.2,
                            rotation_range=0.2,
                            horizontal_flip=True,
                            rescale=1/255)
```

```python
gen_1=data_gen.flow_from_directory('/kaggle/input/agriculture-crop-images/crop_images',target_size=(224,224),
                                  batch_size=64)
```

```python
gen_2=data_gen.flow_from_directory('/kaggle/input/agriculture-crop-images/kag2',target_size=(224,224),
                                  batch_size=64)
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
vgg_model=VGG16(include_top=False,input_shape=(224,224,3))
```

```python
for layer in vgg_model.layers:
    layer.trainable=False
    
```

```python
vgg_model.summary()
```

```python
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.models import Model
```

```python
flatten_layer=Flatten()(vgg_model.output)
```

```python
last_layer=Dense(5,activation='softmax')(flatten_layer)
```

```python
model=Model(inputs=vgg_model.input,outputs=last_layer)
model.compile(loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(gen_1,epochs=15)
```

```python
model.fit(gen_2,epochs=15)
```

```python
model.save('tl_crop.h5')
```
