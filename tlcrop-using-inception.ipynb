```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_gen=ImageDataGenerator(zoom_range=0.2,
                           horizontal_flip=True,
                           rotation_range=0.2,
                           rescale=1/255)
```

```python
gen1=data_gen.flow_from_directory('/kaggle/input/agriculture-crop-images/crop_images',
                                 target_size=(224,224),batch_size=32)
```

```python
gen2=data_gen.flow_from_directory('/kaggle/input/agriculture-crop-images/kag2',
                                 target_size=(224,224),batch_size=32)
```

```python
gen1.class_indices
```

```python
gen2.class_indices
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```python
inception_model=InceptionV3(include_top=False,weights='imagenet',
                           input_shape=(224,224,3))
```

```python
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
```

```python
model.add(inception_model)
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.models import Model
```

```python
flatten_layer=Flatten()(model.output)
last_layer=Dense(5,activation='softmax')(flatten_layer)
```

```python
model=Model(inputs=model.input,outputs=last_layer)
model.compile(loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(gen1,epochs=10)
```

```python
model.fit(gen2,epochs=10)
```

```python
model.save('TLcrop_using_inception.h5')
```
